```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PowerTransformer

# Create the data with outliers
data = np.random.normal(loc=10, scale=5, size=1000)
outliers = np.random.normal(loc=50, scale=10, size=20)
data = np.concatenate([data, outliers])

# Convert data to a pandas DataFrame
df = pd.DataFrame(data, columns=['value'])

# Apply Box-Cox transformation to the data
pt = PowerTransformer(method='box-cox', standardize=True)
df['value_transformed'] = pt.fit_transform(df[['value']])

# Print the mean and standard deviation of the transformed data
print(df['value_transformed'].mean())
print(df['value_transformed'].std())
```

```python
import numpy as np
import pandas as pd

#create the data outliers
data=np.random.normal(loc=10,scale=5,size=1000)
outliers=np.random.normal(loc=50,scale=10,size=20)
data=np.concatenate([data,outliers])
df=pd.DataFrame(data,columns=['value'])

import matplotlib.pyplot as plt
import matplotlib_inline
import seaborn as sns
sns.set_style(style='darkgrid')

sns.boxplot(data=df,x='value')
plt.title('Boxcox plot of the only features in the sample data')
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer

power_transform=PowerTransformer(method='box-cox',standardize=True)
df['transformed_value']=power_transform.fit_transform(df[['value']])

# Print the mean and standard deviation of the transformed data
print(df['value_transformed'].mean())
print(df['value_transformed'].std())
print(20*'*')

sns.boxplot(data=df,x='transformed_value')
plt.title('Boxcox plot boxcox transformed features')
plt.show()
```

The box-cox transformation can be applied only to positive values.

- We apply the Box-Cox transformation using the PowerTransformer class from scikit-learn. We specify the method argument as 'box-cox' and the standardize argument as True, which applies the Box-Cox transformation and standardizes the variables to have zero mean and unit variance.
- Finally, we print the mean and standard deviation of the transformed data using the mean and std functions of the pandas DataFrame.
- Note that in real-world scenarios, the Box-Cox transformation should be applied after removing or treating the outliers. In this example, we kept the outliers to demonstrate the effect of the transformation on non-normal data.

```python
# Create the data with outliers
data = np.random.exponential(scale=1, size=1000)
outliers = np.random.exponential(scale=10, size=20)
data = np.concatenate([data, outliers])

# Convert data to a pandas DataFrame
df = pd.DataFrame(data, columns=['value'])

# Shift the data to make it positive
df['value'] = df['value'] - df['value'].min() + 1

sns.boxplot(data=df,x='value')
plt.title('Boxcox plot of the new positive sample data')
plt.show()
```

```python
# Apply Box-Cox transformation to the data
pt = PowerTransformer(method='box-cox', standardize=True)
df['value_transformed'] = pt.fit_transform(df[['value']])

# Print the mean and standard deviation of the transformed data
print(df['value_transformed'].mean())
print(df['value_transformed'].std())
print(20*'*')

sns.boxplot(data=df,x='value_transformed')
plt.title('Boxcox plot boxcox transformed and standardized feature')
plt.show()
```
